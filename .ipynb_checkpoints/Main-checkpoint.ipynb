## Data Description ##

There are four datasets that contain processed angiography data on patients in various clinics in 1988. Applying a model derived from test results of 303 patients undergoing angiography at the Cleveland Clinic in Cleveland, Ohio to estimate probabilities of angiographic coronary disease. 

Note: The data for each location were initially in a .data format; for the purposes of the project, they will be converted to a .csv format, modified to contain the data’s

Before cleaning the data, there are 14 variables in each dataset. These include:
* `age`: (years)
* `sex`: (male = 1 ; female = 0)
* `cp`: chest pain type (typical angina = 1 ; atypical angina = 2 ; non-anginal pain = 3 ; asymptomatic = 4)
* `trestbps`: resting blood pressure (mmHg) (upon admission to hospital)
* `chol`: serum cholesterol
* `fbs`: fasting blood sugar > 120 mg/dl
* `restecg`: resting electrocardiographic results (1-3)
* `thalach`: maximum heart rate achieved
* `exang`: exercise induced angina (0/1)
* `oldpeak`: ST depression induced by exercise relative to rest
* `slope`: the slope of the peak exercise ST segment (1-3)
* `ca`: number of major vessels (0-3) colored by flourosopy
* `thal`: (3 = normal; 6 = fixed defect; 7 = reversable defect)
* `num`: diagnosis of heart disease (Value 0: <50% diameter narrowing ; Value = 1: >50% diameter narrowing)
* `location`: Four locations: Cleveland, Switzerland, Hungary, VA

The variables age, sex, cp, restecg, and num are `int` variables whereas trestbps, chol, fbs, thalach, exang, oldpeak, slope, ca, thal are `chr` variables. These variables were the ones chosen for all published experiments, hence they are seen as significant factors in studies and kept for the datasets. The variables filtered out are `fbs`, `exang`, `oldpeak`, `slope`, `ca`, and `thal`. These variables contain many missing values. Some variables are binary such as `sex` and `num`. A new variable named `location` was added to specify where the observation was observed. The variable `num` was modified such that any value greater than 1 becomes 1 since all values >= 1 result in the same diagnosis. This was made for better calculation and easier understanding for the reader.
The number of observations for each location is:
* `Cleveland`: 303
* `Switzerland`: 123
* `Hungarian`: 294
* `VA`: 200

**Total**: 920

The distribution amongst how many datapoints each dataset contributes is not equal. Givin that the Cleveland database is the only one that has been used for ML research, it is understandable that it covers the most points between the four sets.

# Assignment 2: Visualization #

One issue that may arise is that there will be multicollinearity between input variables. We will generate a heatmap of the correlation matrix between potential input variables to see if there is any correlation amongst them. We check this because a high correlation in input variables affects the least squares estimates of the linear regression coefficients, standard errors, and significance (p-value). By identifying any correlated variables, we can use Variance Inflation Factor (VIF) to determine which inputs we shouldn't account for. To explore more variables, we will also include `oldpeak` which was previously filtered out.

We begin by removing the factor variables and the response variable from the dataset. We also converted `trestbps`, `thalach`, `oldpeak` and `chol` from `chr` variables to `dbl` variables.

```r
heart_dat <- rbind(Cleveland, Hungarian, Switzerland, VA) |>
filter(!(location == "?" | age == "?" | sex == "?" | cp == "?" | trestbps == "?" | chol == "?" | oldpeak == "?" | restecg == "?" | thalach == "?"|  num == "?"), location != "Switzerland") |>
mutate(chol = as.numeric(chol), trestbps = as.numeric(trestbps), thalach = as.numeric(thalach), oldpeak = as.numeric(oldpeak)) |>
dplyr::select(-c(num, cp, location, restecg, sex, fbs, restecg, exang, slope, ca, thal)) %>%
  mutate_if(is.numeric, round, 2)

head(heart_dat)
tail(heart_dat)
```

Table 2. Filtered Data for Heatmap

We create a correlation table, generating the pairwise correlations between the potential input variables. `var1` is the first input variable, `var2` is the second input variable, and `corr` is the correlation between them.

```r
corr_matrix_heart <- heart_dat %>%  
  cor() %>%
  as.data.frame() %>%
  rownames_to_column("var1") %>%
  pivot_longer(-var1, names_to = "var2", values_to = "corr")
head(corr_matrix_heart)
```

Table 3. Melted Correlation Table

Using the melted correlation table, we can visualize the heatmap matrix below. 

```r
plot_corr_matrix_heart <- corr_matrix_heart %>%
  ggplot(aes(var1, var2)) +
  geom_tile(aes(fill = corr), color = "white") +
  scale_fill_distiller("Correlation Coefficient \n",
    palette = "YlOrRd",
    direction = 1, limits = c(-1, 1)
  ) +
  labs(x = "", y = "") +
  ggtitle("Heatmap of Input Variables to Estimate Heart Disease") +
  theme_minimal() +
  theme(
    axis.text.x = element_text(
      angle = 45, vjust = 1,
      size = 18, hjust = 1
    ),
    axis.text.y = element_text(
      vjust = 1,
      size = 18, hjust = 1
    ),
    legend.title = element_text(size = 18, face = "bold"),
    legend.text = element_text(size = 18),
    legend.key.size = unit(2, "cm")
  ) +
  coord_fixed() +
  geom_text(aes(var2, var1, label = round(corr, 2)), color = "black", size = 6)
plot_corr_matrix_heart
```

### Why is this method appropriate? Which assumptions are required? ###

My question is predicting the diagnosis of heart disease based on continuous and categorical variables such as age, sex, location, resting blood pressure, chest pain type, maximum heart rate, and cholesterol level. The method I propose is binary logistic regression. This method is appropriate because the response variable `num` is binary. Logistic regression uses the log of the odds ratio, which is the ratio between the proportion of success to the proportion of failures. We can generate the model coefficients and their confidence intervals, evaluate the goodness-of-fit for our model, and adjust for overdispersion.  

The assumptions of this regression are that the response must be binary or binomial, the observations must be independent, the variance structure’s highest variability is when the proportion is 0.5, and the logarithm of the odds ratio has to be a linear function of the independent variable.

### Limitations ###

One limitation of this method is that when using a complete case analysis, the variables that are removed because of missing values could contain data that varies from the data we analyzed, which causes bias. In our dataset, we have access to the databases of all four locations, whereas previous studies have only used the Cleveland database, which means a reduced sample size.  Another limitation could be an inaccurate definition of our variables. For example, a high or low fasting blood sugar (`fbs`), could be a fluctuation due to unhealthy habits such as sleeping late or skipping meals, and not necessarily related to heart disease. Another example is that the resting blood pressure (`trestbp`) is only taken upon admission to the hospital, leaving us with no other blood pressure data. We can also assume patients first being brought into the hospital will be sicker, and could have a more fluctuated blood pressure. There may be significant variation in the sample selection and response caused by the administrators that collected the data. 

## Assignment 4: Implementation of Logistic Regression ##

```r
log_heart <- heart |> 
mutate(num = if_else(num == "1", 1, 0)) |> #Turn num into a double 
dplyr::select(location, age, sex, cp, trestbps, chol, restecg, thalach, num) |>
filter(!(location == "?" | age == "?" | sex == "?" | cp == "?" | thalach == "?" | trestbps == "?" | chol == "?" | restecg == "?" | num == "?")) |>
mutate(sex = as.factor(sex), cp = as.factor(cp), restecg = as.factor(restecg), location = as.factor(location), trestbps = as.numeric(trestbps), chol = as.numeric(chol))|>
filter(location != "Switzerland") #too many missing chol variables in Switzerland
#n = 712


head(log_heart)
```

```r
heart_diagnosis_log_model <- 
    glm(
        formula = num ~ age + sex + chol + trestbps + thalach + cp + restecg,
        data = log_heart,
        family = binomial)

heart_diagnosis_binary_log  <-
    tidy(heart_diagnosis_log_model) %>% 
    mutate(exp.estimate = exp(estimate)) %>% 
    mutate_if(is.numeric, round, 3)

heart_diagnosis_binary_log 
```
